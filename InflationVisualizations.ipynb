```python
fig, ax = plt.subplots(figsize=(15, 10))
plt.plot(inf['DATE'], inf['CPIFABSL'],'o', alpha = 1.0)
plt.xlabel('date', fontsize=20)
plt.xticks(rotation=45)
plt.xticks(np.arange(0,52))
plt.locator_params(axis='x', nbins=10)
plt.ylabel('CPIFABSL', fontsize=20)
plt.title("CPIFABSL over Time", fontsize=25)
```

```python
inf['CHANGE'] = inf['CPIFABSL']-inf['CPIFABSL'].shift(1)
```

```python
fig, ax = plt.subplots(figsize = (15, 10))
plt.plot(inf['DATE'], inf['CHANGE'], 'o', alpha = 0.7)

plt.xlabel('date')
plt.xticks(rotation=90)

plt.ylabel('change in CPI')
```

```python
fig, ax = plt.subplots(figsize = (15, 10))
plt.plot(inf['DATE'], inf['CHANGE'], 'o', alpha = 0.7)

plt.xlabel('date')
plt.xticks(rotation=90)


plt.ylabel('change in CPI')
```

```python
inf['CHANGE'].rolling(window=30).mean().plot()
```

```python
inf['CHANGE_7'] = inf['CHANGE'].rolling(7).mean()
```

```python
inf['CPIFABSL_7'] = inf['CPIFABSL'].rolling(7).mean()
```

```python
inf.head(20)
```

```python
fig, ax = plt.subplots(figsize = (15, 10))
plt.plot(inf['DATE'], inf['CPIFABSL'], 'o', alpha = 0.7, label = 'CPIFABSL')
plt.plot(inf['DATE'], inf['CPIFABSL_7'], alpha = 0.9, label = '7 day rolling average of CPIFABSL')
plt.xlabel('date')
plt.xticks(rotation=90)
plt.ylabel('CPIFABSL')
```

```python
fig, ax = plt.subplots(figsize = (15, 10))
plt.plot(inf['DATE'], inf['CHANGE'], 'o', alpha = 0.7)
plt.plot(inf['DATE'], inf['CHANGE_7'], alpha = 0.9, label = '7 day rolling average of change in CPIFABSL')
plt.xlabel('date')
plt.xticks(rotation=90)


plt.ylabel('change in CPI')
```

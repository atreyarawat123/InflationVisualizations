```python
import pandas as pd
import numpy as np 
import matplotlib.pylab as plt
```

```python
pce = pd.read_csv("data/PCE.csv")
```

```python
fig, ax = plt.subplots(figsize=(15, 10))
plt.plot(pce['DATE'], pce['PCE'],'o', alpha = 1.0)
plt.xlabel('date', fontsize=20)
plt.xticks(rotation=45)
plt.xticks(np.arange(0,500))
plt.locator_params(axis='x', nbins=20)
plt.ylabel('PCE', fontsize=20)
plt.title("PCE over Time", fontsize=25);
```

```python
pce['PCHANGE'] = pce['PCE'].pct_change()
fig, ax = plt.subplots(figsize = (15, 10))
plt.plot(pce['DATE'], pce['PCHANGE'], 'o', alpha = 1.0)
plt.axhline(y=0.0, linestyle='--', color = 'black', alpha = 0.7)
plt.xlabel('date', fontsize=20)
plt.xticks(rotation=45)
plt.xticks(np.arange(0,500))
plt.locator_params(axis='x', nbins=20)
plt.ylabel('percent change in PCE', fontsize=20)
plt.title("monthly percent change in PCE over Time", fontsize=25);
```

```python
pce['PACHANGE'] = pce['PCE'].pct_change(periods = 12)
fig, ax = plt.subplots(figsize = (15, 10))
plt.plot(pce['DATE'], pce['PACHANGE'], 'o', alpha = 1.0)
plt.axhline(y=0.0, linestyle='--', color = 'black', alpha = 0.7)
plt.xlabel('date', fontsize=20)
plt.xticks(rotation=45)
plt.xticks(np.arange(0,500))
plt.locator_params(axis='x', nbins=20)
plt.ylabel('percent change in PCE', fontsize=20)
plt.title("annual percent change in PCE over Time", fontsize=25);
```

```python
pce['CHANGE'] = pce['PCE'].diff()
fig, ax = plt.subplots(figsize = (15, 10))
plt.plot(pce['DATE'], pce['CHANGE'], 'o', alpha = 1.0)
plt.axhline(y=0.0, linestyle='--', color = 'black', alpha = 0.7)
plt.xlabel('date', fontsize=20)
plt.xticks(rotation=45)
plt.xticks(np.arange(0,500))
plt.locator_params(axis='x', nbins=20)
plt.ylabel('change in PCE', fontsize=20)
plt.title("absolute change in PCE over Time", fontsize=25);
```

```python
import pandas as pd
import numpy as np 
import matplotlib.pylab as plt
inf = pd.read_csv("../data/inflation.csv")
```

```python
fig, ax = plt.subplots(figsize=(15, 10))
plt.plot(inf['DATE'], inf['CPIFABSL'],'o', alpha = 1.0)
plt.xlabel('date', fontsize=20)
plt.xticks(rotation=45)
plt.xticks(np.arange(0,500))
plt.locator_params(axis='x', nbins=20)
plt.ylabel('CPI', fontsize=20)
plt.title("CPI over Time", fontsize=25)
```

```python
inf['PCHANGE'] = inf['CPIFABSL'].pct_change()
fig, ax = plt.subplots(figsize = (15, 10))
plt.plot(inf['DATE'], inf['PCHANGE'], 'o', alpha = 1.0)
plt.axhline(y=0.0, linestyle='--', color = 'black', alpha = 0.7)
plt.xlabel('date', fontsize=20)
plt.xticks(rotation=45)
plt.xticks(np.arange(0,500))
plt.locator_params(axis='x', nbins=20)
plt.ylabel('percent change in CPI', fontsize=20)
plt.title("monthly percent change in CPI over Time", fontsize=25);
```

```python
inf['PACHANGE'] = inf['CPIFABSL'].pct_change(periods = 12)
fig, ax = plt.subplots(figsize = (15, 10))
plt.plot(inf['DATE'], inf['PACHANGE'], 'o', alpha = 1.0)
plt.axhline(y=0.0, linestyle='--', color = 'black', alpha = 0.7)
plt.xlabel('date', fontsize=20)
plt.xticks(rotation=45)
plt.xticks(np.arange(0,500))
plt.locator_params(axis='x', nbins=20)
plt.ylabel('percent change in CPI', fontsize=20)
plt.title("annual percent change in CPI over Time", fontsize=25);
```

```python
inf['CHANGE'] = inf['CPIFABSL'].diff()
fig, ax = plt.subplots(figsize = (15, 10))
plt.plot(inf['DATE'], inf['CHANGE'], 'o', alpha = 1.0)
plt.axhline(y=0.0, linestyle='--', color = 'black', alpha = 0.7)
plt.xlabel('date', fontsize=20)
plt.xticks(rotation=45)
plt.xticks(np.arange(0,500))
plt.locator_params(axis='x', nbins=20)
plt.ylabel('change in CPI', fontsize=20)
plt.title("absolute change in CPI over Time", fontsize=25);
```

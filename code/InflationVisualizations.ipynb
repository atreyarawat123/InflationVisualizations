```python
import pandas as pd
import numpy as np 
import matplotlib.pylab as plt
```

```python
inf = pd.read_csv("inflation.csv")
```

```python
inf.head(10)
```

```python
inf.sort_values(by = 'CPIFABSL', ascending = False)[:5]
```

```python
fig, ax = plt.subplots(figsize=(15, 10))
plt.plot(inf['DATE'], inf['CPIFABSL'],'o', alpha = 1.0)
plt.xlabel('date', fontsize=20)
plt.xticks(rotation=45)
plt.xticks(np.arange(0,52))
plt.locator_params(axis='x', nbins=10)
plt.ylabel('CPIFABSL', fontsize=20)
plt.title("CPIFABSL over Time", fontsize=25)
```

```python
inf['CPIFABSL_7'] = inf['CPIFABSL'].rolling(7).mean()
```

```python
fig, ax = plt.subplots(figsize = (15, 10))
plt.plot(inf['DATE'], inf['CPIFABSL'], 'o', alpha = 1.0, label = 'CPIFABSL')
plt.plot(inf['DATE'], inf['CPIFABSL_7'], alpha = 1.0, label = '7 day rolling average of CPIFABSL')
plt.xlabel('date', fontsize=20)
plt.xticks(rotation=45)
plt.xticks(np.arange(0,52))
plt.locator_params(axis='x', nbins=10)
plt.ylabel('CPIFABSL', fontsize = 20)
plt.title("CPI 7 day rolling average", fontsize=25)
plt.legend()
```

```python
inf['CHANGE'] = inf['CPIFABSL']-inf['CPIFABSL'].shift(1)
```

```python
fig, ax = plt.subplots(figsize = (15, 10))
plt.plot(inf['DATE'], inf['CHANGE'], 'o', alpha = 1.0)
plt.xlabel('date', fontsize=20)
plt.xticks(rotation=45)
plt.xticks(np.arange(0,52))
plt.locator_params(axis='x', nbins=10)
plt.ylabel('change in CPI', fontsize = 20)
plt.title("change in cpi over time", fontsize=25)
```

```python
inf['CHANGE_7'] = inf['CHANGE'].rolling(7, min_periods=1).mean()
fig, ax = plt.subplots(figsize = (15, 10))
plt.axhline(y=0.0, linestyle='--', color = 'black', alpha = 0.7)
plt.plot(inf['DATE'], inf['CHANGE'], 'o', alpha = 1.0, label = 'CHANGE IN CPI')
plt.plot(inf['DATE'], inf['CHANGE_7'], alpha = 1.0, label = '7 day rolling average of change in CPI')
plt.xlabel('date', fontsize=20)
plt.xticks(rotation=45)
plt.xticks(np.arange(0,52))
plt.locator_params(axis='x', nbins=10)
plt.ylabel('change in CPI', fontsize = 20)
plt.title("change in cpi over time", fontsize=25)
plt.legend()
```

```python
fig, ax = plt.subplots(figsize = (15, 10))
plt.plot(inf['CPIFABSL'], inf['CHANGE'], 'o', alpha = 1.0)
plt.xlabel('cpi', fontsize=20)
plt.xticks(rotation=45)
plt.xticks(np.arange(0,52))
plt.locator_params(axis='x', nbins=10)
plt.ylabel('change in CPI', fontsize = 20)
plt.title("cpi vs change in cpi", fontsize=25)
```

```python
import pandas as pd
import numpy as np 
import matplotlib.pylab as plt
```

```python
pcea = pd.read_csv("PCEA.csv")
```

```python
fig, ax = plt.subplots(figsize=(15, 10))
plt.plot(pcea['DATE'], pcea['PCE'],'o', alpha = 1.0)
plt.xlabel('date', fontsize=20)
plt.xticks(rotation=45)
plt.xticks(np.arange(0,52))
plt.locator_params(axis='x', nbins=10)
plt.ylabel('PCE', fontsize=20)
plt.title("PCE over Time", fontsize=25)
```

```python
pcea['CHANGE'] = pcea['PCE']-pcea['PCE'].shift(1)
```

```python
fig, ax = plt.subplots(figsize = (15, 10))
plt.plot(pcea['DATE'], pcea['CHANGE'], 'o', alpha = 1.0)
plt.xlabel('date', fontsize=20)
plt.xticks(rotation=45)
plt.xticks(np.arange(0,52))
plt.locator_params(axis='x', nbins=10)
plt.ylabel('change in PCE', fontsize=20)
plt.title("change in PCE over Time", fontsize=25)
```

```python
fig, ax = plt.subplots(figsize = (15, 10))
plt.plot(pcea['PCE'], pcea['CHANGE'], 'o', alpha = 1.0)
plt.xlabel('PCE', fontsize=20)
plt.xticks(rotation=45)
plt.xticks(np.arange(0,52))
plt.locator_params(axis='x', nbins=10)
plt.ylabel('change in PCE', fontsize=20)
plt.title("PCE vs change in PCE", fontsize=25)
```

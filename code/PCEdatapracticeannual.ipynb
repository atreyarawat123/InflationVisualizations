```python
import pandas as pd
import numpy as np 
import matplotlib.pylab as plt
```

```python
pcea = pd.read_csv("../data/PCEA.csv")
```

```python
fig, ax = plt.subplots(figsize=(15, 10))
plt.plot(pcea['DATE'], pcea['PCE'],'o', alpha = 1.0)
plt.xlabel('date', fontsize=20)
plt.xticks(rotation=45)
plt.xticks(np.arange(0,52))
plt.locator_params(axis='x', nbins=10)
plt.ylabel('PCE', fontsize=20)
plt.title("PCE over Time", fontsize=25)
```

```python
pcea['PCE_7'] = pcea['PCE'].rolling(7).mean()
fig, ax = plt.subplots(figsize = (15, 10))
plt.plot(pcea['DATE'], pcea['PCE'], 'o', alpha = 1.0, label = 'PCE')
plt.plot(pcea['DATE'], pcea['PCE_7'], alpha = 1.0, label = '7 day rolling average of PCE')
plt.axhline(y=0.0, linestyle='--', color = 'black', alpha = 0.7)
plt.xlabel('date', fontsize=20)
plt.xticks(rotation=45)
plt.xticks(np.arange(0,52))
plt.locator_params(axis='x', nbins=10)
plt.ylabel('PCE', fontsize = 20)
plt.title("PCE 7 day rolling average", fontsize=25)
plt.legend()
```

```python
pcea['CHANGE'] = pcea['PCE']-pcea['PCE'].shift(1)
```

```python
fig, ax = plt.subplots(figsize = (15, 10))
plt.plot(pcea['DATE'], pcea['CHANGE'], 'o', alpha = 1.0)
plt.axhline(y=0.0, linestyle='--', color = 'black', alpha = 0.7)
plt.xlabel('date', fontsize=20)
plt.xticks(rotation=45)
plt.xticks(np.arange(0,52))
plt.locator_params(axis='x', nbins=10)
plt.ylabel('change in PCE', fontsize=20)
plt.title("change in PCE over Time", fontsize=25)
```

```python
pcea['CHANGE_7'] = pcea['CHANGE'].rolling(7, min_periods=1).mean()
fig, ax = plt.subplots(figsize = (15, 10))
plt.axhline(y=0.0, linestyle='--', color = 'black', alpha = 0.7)
plt.plot(pcea['DATE'], pcea['CHANGE'], 'o', alpha = 1.0, label = 'CHANGE IN PCE')
plt.plot(pcea['DATE'], pcea['CHANGE_7'], alpha = 1.0, label = '7 day rolling average of change in PCE')
plt.xlabel('date', fontsize=20)
plt.xticks(rotation=45)
plt.xticks(np.arange(0,52))
plt.locator_params(axis='x', nbins=10)
plt.ylabel('change in PCE', fontsize = 20)
plt.title("change in pce over time", fontsize=25)
plt.legend()
```

```python
fig, ax = plt.subplots(figsize = (15, 10))
plt.plot(pcea['PCE'], pcea['CHANGE'], 'o', alpha = 1.0)
plt.axhline(y=0.0, linestyle='--', color = 'black', alpha = 0.7)
plt.xlabel('PCE', fontsize=20)
plt.xticks(rotation=45)
plt.xticks(np.arange(0,52))
plt.locator_params(axis='x', nbins=10)
plt.ylabel('change in PCE', fontsize=20)
plt.title("PCE vs change in PCE", fontsize=25)
```
